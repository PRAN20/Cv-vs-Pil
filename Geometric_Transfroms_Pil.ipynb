**Importing Header Files**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as ps
```

**Declaring Functions**

```python
def conc_image (image1, image2, title_1 = "Original Image", title_2 = "New Image"):
  plt.figure(figsize=(5,5))
  plt.subplot(1,2,1)
  plt.title(title_1)
  plt.imshow(image1, cmap= "gray")

  plt.subplot(1,2,2)
  plt.title(title_2)
  plt.imshow(image2, cmap= "gray")
  plt.show()
```

# **Geometric Transformation**

```python
from PIL import Image
image = Image.open("image.jpg")
plt.imshow(image)
plt.show()
```

```python
width, height = image.size
new_width = 2 * width
new_height = height
new_image = image.resize((new_width, new_height))
conc_image(image, new_image)
```

```python
width, height = image.size
new_width = width
new_height = 2*height
new_image = image.resize((new_width, new_height))
plt.figure(figsize=(10,10))
conc_image(image, new_image)
```

```python
width, height = image.size
new_width = 2* width
new_height = 2*height
new_image = image.resize((new_width, new_height))
conc_image(image, new_image)
```

```python
width, height = image.size
new_width = width // 2
new_height = height // 2
new_image = image.resize((new_width, new_height))
conc_image(image, new_image)
```

# **Rotation**

```python
imag1 = Image.open("pin.png")
plt.figure(figsize=(10,10))
plt.imshow(imag1)
plt.show()
```

```python
theta = 45
new_image = imag1.rotate(theta)
plt.figure(figsize=(10,10))
conc_image(imag1, new_image)
```
